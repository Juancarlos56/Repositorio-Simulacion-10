### Mapas de Calor

#### Mapa calor de FIFA

```python

corrEnfermedades = fifa_df.corr()
corrEnfermedades = corrEnfermedades.round(2)
plt.rc('font', size = 10)
plt.subplots(figsize=(15, 12))
dropSelf = np.zeros_like(corrEnfermedades)
dropSelf[np.triu_indices_from(dropSelf)] = True
sns.set(font_scale=1)
mapaCalor = sns.heatmap(corrEnfermedades, square=True, mask=dropSelf, linewidths=.5, annot=False)
```

#### Mapa calor de empleo

```python

corrEnfermedades = empleo_df.corr()
corrEnfermedades = corrEnfermedades.round(2)
plt.rc('font', size = 10)
plt.subplots(figsize=(15, 12))
dropSelf = np.zeros_like(corrEnfermedades)
dropSelf[np.triu_indices_from(dropSelf)] = True
sns.set(font_scale=1)
mapaCalor = sns.heatmap(corrEnfermedades, square=True, mask=dropSelf, linewidths=.5, annot=False)
```
