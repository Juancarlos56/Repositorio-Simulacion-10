### Distribución Triangular

El nombre de esta distribución viene dado por la forma de su función de densidad. Este
modelo proporciona una primera aproximación cuando hay poca información disponible, de
forma que sólo se necesita conocer el mínimo (valor pesimista), el máximo (valor optimista)
y la moda (valor más probable). Estos tres valores son los parámetros que caracterizan a la
distribución triangular y se denotan por a, b y c, respectivamente.

Formula: 

image.png

Campo de variación:


$$a\leq X \leq b $$

Parámetros:

- a: mínimo, $-\infty < a < \infty$
- c: moda, $-\infty < c < \infty$  con  $a \leq c \leq b$
- b: máximo, $-\infty < b < \infty$ con $a < b$

### Funcion de Densidad de Probabilidad PDF y Función de Densidad Acumulada CDF

```python
### Codigo de funcion transformada inversa de distribucion triangular

def transformadaInversa(num_aleatorios, valorMinimo, valorMedio, valorMaximo):
  valores = []
  intervalo = (valorMedio - valorMinimo)/(valorMaximo- valorMinimo)
  for u in num_aleatorios:
    if(0<u and u<intervalo):
      r = valorMinimo + math.sqrt((valorMaximo-valorMinimo)*(valorMedio-valorMinimo)* u)
    if(intervalo<=u and u<1):
      r = valorMaximo - math.sqrt((valorMaximo-valorMinimo)*(valorMaximo-valorMinimo)*(1-u))
    valores.append(r)
  return valores
```

```python
from scipy.stats import uniform, triang
import math
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

### Generate exponential distributed random variables given the mean 
### and number of random variables
def triangular_inverse_trans(numerosRandomLibreary,numeroAleatorios=1, valorMaximo=1, valorMedio = 0.5, valorMinimo=0):
    #Generacion de numeros aleatorios:
    #U = uniform.rvs(size=numeroAleatorios, random_state=2023)
    U = numerosRandomLibreary
    #normalizando valores entre 0-1, datos de entreda
    #Orden, menor, medio, mayor
    list = np.array([valorMinimo,valorMedio,valorMaximo]).reshape(-1, 1)
    scaler = MinMaxScaler(feature_range=(0, 1))
    normalizedlist=scaler.fit_transform(list)

    #Transformada Inversa generada
    X= transformadaInversa(U, normalizedlist[0][0],normalizedlist[1][0],normalizedlist[2][0])
    print("Cantidad de valores obtenidos en X=",len(X))
    
    #Transformada inversa de spicy
    actual = triang.rvs(normalizedlist[1][0], size=100 , random_state=2023)
    print("Cantidad de valores obtenidos en actual=",len(actual))
    


    ##Plot Histograma
    plt.figure(figsize=(12,9)) 
    plt.hist(X, bins=50, alpha=0.5, label="Generated r.v.")
    plt.hist(actual, bins=50, alpha=0.5, label="Actual r.v.")
    
    plt.title("Generada vs Actual Trinagular Random Variables")
    plt.legend()
    plt.show()

    return X
```

```python
resultado=triangular_inverse_trans(numerosRandomLibreary, numeroAleatorios=100,valorMaximo=100, valorMedio=78.52)
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(triang.ppf(0.0, 0.7852),
                triang.ppf(1.0, 0.7852), 100)

ax.plot(x, triang.pdf(x, 0.7852),'-', lw=5, alpha=0.6, label='PDF')
ax.plot(x, triang.cdf(x, 0.7852),'-', lw=5, alpha=0.6, label='CDF')
r = triang.rvs(0.7852, size=100, random_state=2023)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```
