### Obtener las empresas que mayor patrimonio tienen (10).

```python
df['Total Patrimonio (7) 2020'].rank(method='max')
```

```python
dfTopEmpresas = df[["Nombre Grupo Económico","Total Patrimonio (7) 2020", 'Total Patrimonio (7) 2019']]
dfTopEmpresas
```

```python
dfTopEmpresas['sum'] = dfTopEmpresas.loc[dfTopEmpresas['Total Patrimonio (7) 2020'] >= 0,['Total Patrimonio (7) 2020','Total Patrimonio (7) 2019']].sum(axis=1)
dfTopEmpresas['sum'] = dfTopEmpresas['sum'].div(2)
dfTopEmpresas
```

```python
dfTopEmpresas['default_rank'] = dfTopEmpresas['sum'].rank()
dfTopEmpresas
```

#### Top 10 empresas con mayor patrimonio entre 2019-2020

```python
print("Importancia: ")
dfTopEmpresas.nlargest(10, ['default_rank']) 
```
